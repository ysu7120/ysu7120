```python
n = int(input())
p = []

for i in range(n):
    p.append(list(map(int, input().split())))
for i in range(1, len(p)):
    p[i][0] = min(p[i - 1][1], p[i - 1][2]) + p[i][0] # i번열을 빨간집으로 고르는 경우 (i-1번열에서 파란집과 초록집중 더 작은값을 더함)
    p[i][1] = min(p[i - 1][0], p[i - 1][2]) + p[i][1] # i번열을 파란집으로 고르는 경우 (i-1번열에서 초록집과 빨간집중 더 작은값을 더함)
    p[i][2] = min(p[i - 1][0], p[i - 1][1]) + p[i][2] # i번열을 초록집으로 고르는 경우 (i-1번열에서 빨간집과 파란집중 더 작은값을 더함)
print(min(p[n - 1][0], p[n - 1][1], p[n - 1][2]))
```
