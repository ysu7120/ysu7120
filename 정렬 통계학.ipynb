```python
import sys
from collections import Counter

N=int(sys.stdin.readline())
lst=[]
for i in range(N):
  k=int(sys.stdin.readline())
  lst.append(k)

lst.sort()
print(round(sum(lst)/N))
print(lst[N//2])

from collections import Counter
cnt_lst = Counter(lst).most_common()
if len(cnt_lst) > 1 and cnt_lst[0][1]==cnt_lst[1][1]:
    print(cnt_lst[1][0])
else:
    print(cnt_lst[0][0])

print(max(lst)-min(lst))



```
